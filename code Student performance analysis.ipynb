```python
import pandas as pd

# Load the dataset
df = pd.read_csv("student-dataset.csv")

# Display the first few rows
print(df.head())
```

```python
# Check for missing values
print(df.isnull().sum())

# Handle missing values (you can choose to drop or fill them)
#df = df.fillna(df.mean())  # For numerical columns, filling with the mean value

# Or, for categorical columns, you might use the mode:
# df['engagement_level'] = df['engagement_level'].fillna(df['engagement_level'].mode()[0])

# Check for duplicates
df = df.drop_duplicates()

# Verify the changes
print(df.info())
```

```python
# Summary statistics for numerical columns
print(df.describe())
```

```python
import seaborn as sns #statistical graphics
import matplotlib.pyplot as plt #potting
numerical=df.select_dtypes(include=[np.number]).columns
print(numerical)
# Correlation matrix  linear reltionship between numerical columns
corr_matrix = df[numerical].corr()
print(corr_matrix)

# Plotting the correlation matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()

```

```python
# Create a figure with larger size to fit subplots
plt.figure(figsize=(20, 6))  # Adjust the size as per the number of features

# Loop over each feature to create a subplot
for i, feature in enumerate(numerical):
    plt.subplot(3, 4, i+1)  # 3 rows and 4 columns (adjust as needed)
    sns.histplot(df[feature], bins=30, kde=True, color='skyblue')
    plt.title(f'Distribution of {feature}')  # Title for each histogram

# Adjust layout to avoid overlap
plt.tight_layout()

# Overall Title for the entire plot
plt.suptitle('Distribution of Test Scores', y=1.05, fontsize=16)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assuming df is your dataframe with student data
# We need to extract only the grade/score columns that are numeric

# List of columns that contain grades
grade_columns = ['english.grade', 'math.grade', 'sciences.grade', 'language.grade',
                 'portfolio.rating', 'coverletter.rating', 'refletter.rating']

# Subset the data to contain only the grade columns
student_grades = df[grade_columns]

# Drop columns that have constant values (if any) or NaN values
student_grades = student_grades.loc[:, student_grades.nunique() > 1]

# Plot histograms for each of the remaining grade columns
plt.figure(figsize=(12, 8))
plt.title(f"Visualization of performance report of {student_name}\n \n \n", fontsize=16, fontweight='bold')
for i, column in enumerate(student_grades.columns, 1):
    plt.subplot(2, 4, i)  # 2 rows, 4 columns of subplots (adjust according to the number of columns)
    sns.histplot(student_grades[column], kde=True, color='skyblue', bins=10)
    plt.title(f'Distribution of {column}')
    plt.xlabel('Score')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
